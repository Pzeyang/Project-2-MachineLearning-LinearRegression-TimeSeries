```python
import pandas as pd
```

```python
df = pd.read_csv('best_selling_cars.csv', index_col=0)
df.head()
```

```python
df[['Manufacturer','Model']] = df.Model.str.split(" ", n=1, expand=True) 
```

```python
df.head()
```

```python
cols=df.columns
cols
```

```python
cols = ['Manufacturer', 'Model', 'Month', 'Month_LY', 'Change', 'YTD', 'YTD_LY', 'YTD_Change']
df = df[cols]
df.head()
```

```python
df=df.replace(',','', regex=True)
df=df.sort_values('YTD', ascending=False)
df.head()
```

```python
df[['Month', 'Month_LY','YTD', 'YTD_LY']]=df[['Month', 'Month_LY','YTD', 'YTD_LY']].astype(int)
df.info()
```

```python
mask = df.Manufacturer=='Ram'
df_dodge=df[mask]
df_dodge.head()
```

```python
df.loc[df.Manufacturer == 'Ram', 'Manufacturer'] = 'Dodge'
```

```python
df['Manufacturer'].unique()
```

```python
df_manu = df.groupby('Manufacturer')['YTD', 'YTD_LY'].sum()
#df_manu = df_manu.reset_index()
df_manu = df_manu.sort_values('YTD', ascending=False).reset_index()
df_manu.head()
```
