```python
import pandas as pd
import matplotlib.pyplot as plt
import statsmodels as sm
from statsmodels.tsa.stattools import adfuller
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
from statsmodels.tsa.arima_model import ARIMA
import statistics
import matplotlib.style as style
```

```python
style.available
```

```python
monthly_sentiment_series = pd.read_csv('sales1.csv',
                                     index_col=0, parse_dates=True,
                                     squeeze=True)
#squeeze option will return a series
df_full = monthly_sentiment_series
#print(df_full.tail())
#print('type(df): ', type(df_full.index))
#print('df.dtypes: ', df_full.dtypes)
#print('df.describe() \r\n', df_full.describe())
df_train = df_full.iloc[0:515]
#print('df.describe() \r\n', df_train.describe())
```

#Autoregressive - forecast next timestamp's values by regressing over previous values

Moving average - forecast next timestamp's values by averaging over previous values

ARIMA - useful for non-stationary data, allows us to difference the data. Also includes seasonal differencing parameter for non-stationary data

```python
df_train_diff1 = df_train.diff().fillna(df_train)
df_train_diff2 = df_train_diff1.diff().fillna(df_train_diff1)
print(df_train.tail())
```

```python
style.use('fivethirtyeight')
fig, ax= plt.subplots(figsize=(8,5))
fig.tight_layout()
ax.plot(df_train, label='Car_Sales_Data')
ax.plot(df_train_diff1, linestyle='dashed', label='diff1')
ax.plot(df_train_diff2, linestyle='dotted', label='diff2')
ax.legend(loc='best', frameon=False)
ax.tick_params(axis='both', which='major', labelsize=18)
ax.axhline(y = -20, color = 'black', linewidth = 1.3, alpha = .7)
ax.set_ylabel('Number of Units Sold per Month')
ax.set_xlabel('Year')
plt.title('Monthly Sales (Number of Cars) Figures \r\n for the Automotive Industry')
plt.xticks(rotation=90)
plt.show()
```

## look at ACF and PACF plots

ACF and PACF charts will give us an idea of the number of terms in the ARIMA model.

```python
plot_acf(df_train_diff1, c='orange', label='diff1')
plt.legend(loc='upper right')
plot_acf(df_train_diff2, c='blue', label='diff2')
plt.legend(loc='upper right')
plt.show()
```

```python
plot_pacf(df_train_diff1, c='orange', label='diff1')
plt.legend(loc='upper right')
plot_pacf(df_train_diff2,  label='diff2')
plt.legend(loc='upper right')
plt.show()
```

```python
# blue area - 95% confidence interval. anything outside blue area --> significant correlation.
# any significant spike in lag x in ACF plot --> number of moving average terms.
# any significant spike in lag x in PACF plot --> number of autoregressive terms.
```

```python
ARIMA_model1=ARIMA(df_train, order=(3, 1, 1)).fit(transparams=False)
#Transparams = True means stationary. 
print(ARIMA_model1.summary())
```

```python
ARIMA_forecast = ARIMA_model1.predict('10-01-2018', '12-01-2019', typ='levels')
print('forecast/predictions', ARIMA_forecast)
print(df_full.tail(12))
```

```python
## Transform the differenced X-scale back to original time scale
```

```python
undiff_train_1 = df_train_diff2.cumsum().fillna(df_train_diff2)
undiff_train_2 = undiff_train_1.cumsum().fillna(undiff_train_1)
#print(undiff_train_2.tail(12))
#undiff1 = arima.cumsum().fillna(df_diff1)
print(all(round(df_train, 6)==round(undiff_train_2, 6)))
print('Original Values', df_train.tail())
print('De-Differenced Values', undiff_train_2.tail())
```

```python
predicted_value_series = pd.Series(ARIMA_forecast, index=['2018-10-01', '2018-11-01', '2018-12-01', '2019-01-01',
                                   '2019-02-01', '2019-03-01', '2019-04-01', '2019-05-01',
                                   '2019-06-01', '2019-07-01', '2019-08-01', '2019-09-01',
                                   '2019-10-01', '2019-11-01', '2019-12-01'])
print (predicted_value_series.tail(18))
```

```python
style.use('fivethirtyeight')
df_full_predicted = df_train.iloc[0:513].append(predicted_value_series)
plt.plot(df_full.tail(30), c='blue', label='original')
plt.plot(df_full_predicted.tail(30), linestyle='dashed', c='orange', label='predicted')
plt.xticks(rotation=90)
plt.legend(loc='best', facecolor='none')
plt.show()
```

```python
actual_values_holdout = df_full.iloc[513:528]
predicted_values_holdout = df_full_predicted.iloc[-15:]
prediction_errors = []
for i in range(0,len(actual_values_holdout)):
    err=actual_values_holdout[i]-predicted_values_holdout[i]
    prediction_errors.append(err)
print(prediction_errors)
```

```python
mean_abs_error = statistics.mean(map(abs, prediction_errors))
print(mean_abs_error)
```
